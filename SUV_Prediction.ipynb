```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

from sklearn.neighbors import KNeighborsClassifier

from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv("/content/suv_data.csv")

data
```

```python
data = data.drop(['User ID'],axis=1)
data
```

```python
data.isna().sum()
```

```python
data.dtypes
```

```python
data['Gender'] = data.Gender.map({'Male':0,'Female':1})
data
```

```python
features = data.drop(['Purchased'],axis=1)
target = data['Purchased']
```

```python
x_train,x_test,y_train,y_test = train_test_split(features,target,test_size=0.20,random_state=2)
```

```python
sc = StandardScaler()
x_train = pd.DataFrame(sc.fit_transform(x_train))
x_test = pd.DataFrame(sc.fit_transform(x_test))
```

```python
model_k1 = KNeighborsClassifier(n_neighbors=1)
model_k1.fit(x_train,y_train)

score = model_k1.score(x_test, y_test)
print("{:.3f}".format(score))
```

```python
model_k3 = KNeighborsClassifier(n_neighbors=3)
model_k3.fit(x_train,y_train)

score = model_k3.score(x_test, y_test)
print("{:.3f}".format(score))
```

```python
model_k5 = KNeighborsClassifier(n_neighbors=5)
model_k5.fit(x_train,y_train)

score = model_k5.score(x_test, y_test)
print("{:.3f}".format(score))
```
